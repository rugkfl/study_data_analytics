```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df_Titanic = pd.read_csv('TitanicFromDisaster_train.csv')
df_Titanic
```

## 1. 'Age' 열에서 결측치를 평균 나이로 채우기.

```python
df_age = df_Titanic['Age'].mean()
df_age = round(df_age, 2)
df_age
```

```python
df_Titanic.isnull().sum()
```

```python
df_Titanic_age = df_Titanic['Age']
df_Titanic_age.describe()
df_Titanic['Age'].fillna(df_age)
```

## 2. 'Cabin' 열에 결측치가 있는 행을 찾고, 해당 행들의 평균 'Fare' 계산하기.

```python
df_Titanic['Cabin']
```

```python
df_cabin_na = df_Titanic[df_Titanic.Cabin.isna()]
df_cabin_na_average = df_cabin_na['Fare'].mean()
df_Fare_average = round(df_cabin_na_average, 2)
df_Fare_average
```

## 3. 'Embarked' 열에서 결측치가 있는 행을 찾고, 가장 많이 등장하는 항구로 채우기.


```python
df_Titanic[df_Titanic.Embarked.isna()]
```

```python
df_Titanic.value_counts('Embarked') # 'S'가 가장 많이 등장
```

```python
df_Titanic['Embarked'].fillna('S', inplace=True)
```

## 4. 'Fare' 열에서 평균보다 3 표준편차(mean + 3 * std) 이상 높은 이상치를 가진 행들을 찾기.


```python
df_Titanic['Fare'].describe()
```

```python
Q1 = df_Titanic['Fare'].describe().loc['mean']
Q3 = df_Titanic['Fare'].describe().loc['std']
Q1 = round(Q1, 2)
Q3 = round(Q3, 2)
Q1, Q3
```

```python
bound_max = Q1 + 3 * Q3
bound_max = round(bound_max, 2)
bound_max
```

```python
df_Titanic.query('Fare >= 181.27')
```

## 6. 'Parch' (부모/자녀 수) 열에서 0이 아닌 값을 가진 행들의 평균 'Age' 계산하기.


```python
df_Titanic['Parch']
```

```python
df_Parch_not_zero = df_Titanic.query('Parch != 0')
df_parch_age = df_Parch_not_zero['Age'].mean()
round(df_parch_age, 2)
```

## 7. 'Fare'가 0인 행들을 찾고, 이를 평균 'Fare'로 채우기

```python
df_fare_zero = df_Titanic.query('Fare == 0')  # 평균 fare = df_Fare_average(19.16)
df_fare_zero['Fare'].replace(0, 19.16, inplace=True)
df_fare_zero
```

## 8. 'Pclass' (티켓 등급)가 1등급인 승객들 중 'Cabin' 열의 결측치 비율 계산하기

```python
df_first_class = df_Titanic[df_Titanic['Pclass'] == 1]
df_first_class
```

```python
(df_first_class['Cabin'].isnull().sum()/len(df_first_class['Cabin'])).round(2)
# 결측치의 개수를 총 데이터의 개수로 나눠서 'Cabin'이라는 열의 결측치가 해당 열의 전체 데이터 중 어느 정도의 비율을 차지하는지를 계산
```
