## 연속형/연속형
- 판매된 수량(X)이 각 과일 유형의 단가(Y)와 어떤 관계가 있습니까?
  + 분석 방법: 수량과 가격 사이의 관계의 강도와 방향을 측정하기 위한 피어슨 또는 스피어만


```python
import pandas as pd
df_FS = pd.read_csv('fruitstore_saleshistory.csv')
df_FS
```

```python
df_fruits = df_FS[df_FS['물품중분류'] == '과일']
df_fruits
```

```python
df_fruits[['구매수량', '구매금액']].isnull().sum()
df_FS_amount_sum = df_fruits[['구매수량', '구매금액']].dropna()
df_FS_amount_sum.isnull().sum()
```

```python
df_fruits[['구매수량', '구매금액']].count()
```

```python
df_FS_amount_sum.corr()
```

```python
import scipy.stats as stats
stats.spearmanr(df_FS_amount_sum['구매수량'], df_FS_amount_sum['구매금액'])
```
